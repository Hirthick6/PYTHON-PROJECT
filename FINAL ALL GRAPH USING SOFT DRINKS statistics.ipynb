```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df= pd.read_csv('Soft_drink.csv')
df
```

```python
df.Frequency
```

```python
frequency=np.array(df.Frequency)
frequency
```

```python
total = frequency.sum()
```

```python
total
```

```python
relative_frequency = np.array(frequency/total)
relative_frequency
```

```python
percent_frequency = np.array(relative_frequency*100)
percent_frequency
```

# Frequency

```python
%matplotlib inline
import matplotlib.pyplot as plt
colour = ["blue","green","red","yellow","orange"]
plt.bar(df.Soft_Drinks,frequency,width = 0.5,color = colour)
plt.xlabel("Drink")
plt.ylabel("Frequency")
plt.title("Bar Graph for Frequency")
plt.show
```

# Relative Frequency

```python
plt.bar(df.Soft_Drinks,relative_frequency,width = 0.5,color = colour)
plt.xlabel("Drink")
plt.ylabel("Relative_Frequency")
plt.title("Bar Graph for Relative_Frequency")
plt.show
```

# Percentage Frequency

```python
plt.bar(df.Soft_Drinks,percent_frequency,width = 0.5,color = colour)
plt.xlabel("Drink")
plt.ylabel("Percent_Frequency")
plt.title("Bar Graph for Percent_Frequency")
plt.show
```

# Line Plot Graph

```python
from matplotlib import rcParams
rcParams['figure.figsize']=10,5
plt.plot(df.Frequency)
plt.grid(True,color='b',linestyle=':')
plt.title("line Plot Graph")
plt.xlabel("Drink")
plt.ylabel("Frequency")
```

# Pie-Chart

```python
plt.pie(percent_frequency,labels=df.Soft_Drinks,shadow=True,autopct='%1.1f%%',radius=2.5)
plt.axis("equal")
plt.title("Bar Graph for Percent_Frequency")
plt.show
```

# Histogram

```python
plt.hist(df.Frequency,5,ec="yellow")
plt.xlabel("drinks")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show
```

# Scatter Plot

```python
plt.scatter(df.Soft_Drinks,frequency)
plt.xlabel("drinks")
plt.ylabel("Frequency")
plt.title("Scatter Plot")
plt.show
```
